```python
#I import the libraries i need. 
# Import pandas for data, seaborn and matplotlib for charts.
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
```

```python
#Load the Auto MPG dataset from the web, give column names, and read it into a DataFrame.
 
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
cols = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"]
df = pd.read_csv(url, names=cols, sep='\s+')
```

```python
#Clean the horsepower column by replacing missing values and converting it to numbers.
df["horsepower"] = pd.to_numeric(df["horsepower"].replace("?", pd.NA))
#Change the origin numbers (1, 2, 3) into words: USA, Asia, and Europe.
df["origin"] = df["origin"].map({1:"USA",2:"Asia",3:"Europe"})
```

```python
#Make a bar chart showing how many cars have each number of cylinders.
sns.countplot(x="cylinders", data=df); plt.title("Cylinders Distribution"); plt.show()
#Make a scatterplot showing the relationship between weight and horsepower, colored by origin.
sns.scatterplot(x="weight", y="horsepower", hue="origin", data=df); plt.title("Horsepower vs Weight"); plt.show()
#Make a boxplot showing how MPG changes with the number of cylinders.
sns.boxplot(x="cylinders", y="mpg", data=df); plt.title("MPG by Cylinders"); plt.show()
```
